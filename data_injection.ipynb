```python
# all unique values
import re

all_categories_list=[]
for i in df['category']:
    if type(i)!=str:
        continue
    parts = re.split(r'\s\|\s|,\s|\s&\s', i)
    for j in parts:
        if j.strip() not in all_categories_list:
            all_categories_list.append(j.strip())
```

```python
cate_num=[]
for i in all_categories_list:
    df_=df[df["category"].str.contains(i, na=False)]
    cate_num.append((df_, i))
cate_num.sort(reverse=True)
```

```python
# store categories in a file 
with open("categories.txt", "w") as f:
    for i in all_categories_list:
        f.write(i + "\n")
```
